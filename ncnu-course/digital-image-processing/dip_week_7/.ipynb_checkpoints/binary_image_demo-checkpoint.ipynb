```python
"""
Morphological Transformations
"""

img_j = cv2.imread('j.png',0)
plt.imshow(img_j)
plt.show()
kernel = np.ones((5,5),np.uint8)
```

```python
"""
1. Erosion
"""
erosion = cv2.erode(img_j,kernel,iterations = 1)
plt.imshow(erosion,cmap="gray")
plt.show()
```

```python
"""
2. Dilation
"""
dilation = cv2.dilate(img_j,kernel,iterations = 1)
plt.imshow(dilation,cmap="gray")
plt.show()
```

```python
"""
3. Opening
"""
opening = cv2.morphologyEx(img_j, cv2.MORPH_OPEN, kernel)
plt.imshow(opening,cmap="gray")
plt.show()
```

```python
"""
4. Closing
"""
closing = cv2.morphologyEx(img_j, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing,cmap="gray")
plt.show()
```
